# Kỳ khoản giảm dần

https://www.vietinbank.vn/vn/ca-nhan/cong-cu-tien-ich/cong-cu-tinh-toan/tinh-lich-tra-no-theo-du-no-giam-dan.html

```python
# nhập liệu 
def nhap_lieu():
  so_tien_vay = int(input('Nhập số tiền vay:'))
  ky_han_vay = int(input('Nhập kỳ hạn vay (tháng):'))
  lai_suat_nam = float(input('Nhập lãi suất vay (năm):'))
  return so_tien_vay, ky_han_vay, lai_suat_nam


# bảng dư nợ giảm dần theo PT kỳ khoản giảm dần
def ky_khoan_giam_dan(**kwargs):
  so_tien_vay, ky_han_vay, lai_suat_nam = nhap_lieu()
  lai_suat_thang = lai_suat_nam/12
  df = pd.DataFrame(index = range(0,ky_han_vay + 1), 
                  columns = ['Kỳ', 'Dư_nợ_hiện_tại', 'Nợ_gốc_hàng_tháng', 'Lãi_trả_hàng_tháng', 'Tổng_số_tiền_phải_trả_hàng_tháng'], dtype = 'int')
  df['Kỳ'] = range(0, ky_han_vay + 1)
  df.loc[0,'Dư_nợ_hiện_tại']= so_tien_vay
  df.loc[0,'Nợ_gốc_hàng_tháng']= 0
  df.loc[0,'Lãi_trả_hàng_tháng']= 0
  df.loc[0,'Tổng_số_tiền_phải_trả_hàng_tháng']= 0
  for i in range(1, ky_han_vay+ 1):
    df.loc[i,'Nợ_gốc_hàng_tháng'] = so_tien_vay/ky_han_vay
    df.loc[i,'Dư_nợ_hiện_tại'] = so_tien_vay - i*(so_tien_vay/ky_han_vay)
    df.loc[i,'Lãi_trả_hàng_tháng'] = so_tien_vay*lai_suat_thang*(ky_han_vay+1-i)/ky_han_vay
    df.loc[i,'Tổng_số_tiền_phải_trả_hàng_tháng'] =  so_tien_vay*(ky_han_vay*lai_suat_thang+1-lai_suat_thang*(i-1))/ky_han_vay
  # dòng total 
  df1 = pd.DataFrame({'Kỳ':[0],
                    'Dư_nợ_hiện_tại':[0],
                    'Nợ_gốc_hàng_tháng':[df['Nợ_gốc_hàng_tháng'].sum()],
                    'Lãi_trả_hàng_tháng':[df['Lãi_trả_hàng_tháng'].sum()],
                    'Tổng_số_tiền_phải_trả_hàng_tháng':[df['Tổng_số_tiền_phải_trả_hàng_tháng'].sum()]})

  # 
  df = df.append(df1, ignore_index = True)
  return df.style.format("{:,.0f}")
```

# Dư nợ giảm dần, kỳ khoản cố định 
- Kỳ khoản trả nợ (vốn+lãi) = P * r / ((1-(1+r)^(-n)) = a
- Vốn phải trả kỳ i = a(1+r)^(-(n-i+1))


```python
# nhập liệu 
def nhap_lieu():
  so_tien_vay = int(input('Nhập số tiền vay:'))
  ky_han_vay = int(input('Nhập kỳ hạn vay (tháng):'))
  lai_suat_nam = float(input('Nhập lãi suất vay (năm):'))
  return so_tien_vay, ky_han_vay, lai_suat_nam


# bảng dư nợ giảm dần theo PT kỳ khoản cố định 
def ky_khoan_co_dinh(**kwargs):
  so_tien_vay, ky_han_vay, lai_suat_nam = nhap_lieu()
  lai_suat_thang = lai_suat_nam/12
  ky_khoan_tra_no = (so_tien_vay*lai_suat_thang) / (1-(1+lai_suat_thang)**(-ky_han_vay)) 
  df = pd.DataFrame(index = range(1,ky_han_vay+1), 
                  columns = ['Số kỳ trả', 'Dư nợ đầu kỳ', 'Vốn phải trả', 'Lãi phải trả', 'Vốn + Lãi'], dtype = 'int')
  df['Số kỳ trả'] = range(1, ky_han_vay+1)
  df.loc[1,'Dư nợ đầu kỳ']= so_tien_vay
  for i in range(1, ky_han_vay+1):
    df.loc[i,'Vốn + Lãi'] =  ky_khoan_tra_no
    df.loc[i,'Vốn phải trả'] = ky_khoan_tra_no*(1+lai_suat_thang)**(-(ky_han_vay-i+1))
    df.loc[i+1,'Dư nợ đầu kỳ'] = df.loc[i,'Dư nợ đầu kỳ'] - df.loc[i,'Vốn phải trả']
    df.loc[i,'Lãi phải trả'] = df.loc[i,'Dư nợ đầu kỳ']*lai_suat_thang
  # 
  df = df.iloc[0:ky_han_vay,:]
  return df.style.format("{:,.0f}")
```
