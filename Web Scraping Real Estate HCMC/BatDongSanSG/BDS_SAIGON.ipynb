```python
# Get one page by URL
url = 'https://batdongsansaigon.com.vn/tinh-thanh/tp-ho-chi-minh/?start=0/'
req = Request(url, headers = {'User-Agent': 'Mozilla/5.0'})
driver.get(url) 
driver.title
```

## Get links for 54 pages 

```python
def get_link_by_text(text): 
  """Find link in the page with given text"""
  element = driver.find_element(By.LINK_TEXT, text)
  return element.get_attribute("href")

all_pages = [ get_link_by_text(str(i)) for i in range(1,55)]
all_pages # 54 pages
```

## Get all title of items 

```python
# get all item names 
def get_title(pages):
  res = []
  for ele in pages:  # ele is one of 54 pages 
    driver.get(ele)
    title = driver.find_elements(By.CSS_SELECTOR, '.info-realty')
    sleep(randint(1,5))
    title = [x.text.strip().split('\n')[0] for x in title]
    res = res + title 
  return res 

item_name  = get_title(all_pages)
item_name # 160 item names 
```

```python
all_url = []
for page in all_pages:  
  # use bs4 
  req = Request(page, headers = {'User-Agent': 'Mozilla/5.0'})
  bs = BeautifulSoup(urlopen(page).read(), 'html.parser')
  link = bs.find_all('div', attrs = {'class':'info-realty'}) 
  link = [ x.find_all(['h4', 'a'])[0].find('a')['href'] for x in link]
  all_url = all_url + link 

all_url # 160 url
```

## Get details of content 

```python
# get content 
def get_content(urls): # 160 urls 
  content = []
  for x in urls: 
    req = Request(x, headers = {'User-Agent': 'Mozilla/5.0'})
    driver.get(x)
    ele = driver.find_elements(By.CLASS_NAME,'meta-info-da')
    sleep(randint(1,5))
    ele = [x.text.strip() for x in ele]
    content += ele
  return content  

details  = get_content(all_url)
print(len(details))
details 
```

```python
def get_info(lst):
  item_code = []
  address = []
  property_type = []
  for item in lst: 
    # code 
    id0 = item[0].find(':',0, len(item[0]))
    item_code.append(item[0][id0+1:].strip())
    # address 
    id1  = item[1].find(':',0, len(item[0]))
    address.append(item[1][id1+1:].strip())
    # type 
    id2 = item[2].find(':',0, len(item[0]))
    property_type.append(item[2][id2+1:].strip())
  return  item_code, address, property_type
item_code, address, property_type = get_info(des)
```

```python
def get_price_area(pages):
  res1 = []
  res2 = []
  for page in pages: 
    driver.get(page)
    # price 
    price = driver.find_elements(By.CLASS_NAME,'price')
    sleep(randint(1,5))
    price = [ x.text for x in price]
    res1 += price 
    # get area (diện tích)
    area =  driver.find_elements(By.CLASS_NAME,'area')
    sleep(randint(1,5))
    area =  [x.text for x in area]
    res2 += area 
  return res1, res2 

price, area = get_price_area(all_pages)
```

```python
#post-2614 > div.entry-content > div.realty-infor > p

def get_info_property(urls):
  res = []
  for url in urls: 
    driver.get(url)
    info_property = driver.find_elements(By.CSS_SELECTOR, 'div.realty-infor > p')
    info_property = [x.text for x in info_property]
    res.append(info_property)
  return res 

info_property = get_info_property(all_url)
info_property 
```
